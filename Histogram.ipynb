```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
df_train = pd.read_csv("train.txt", sep="\t", names = ['a', 'b'])
#df_val = pd.read_csv("validation.txt", sep="\t", names = ['a', 'b'])

hashtags_counts = Counter(df_train['b'])
print("Quantidade de hashtags: ", len(hashtags_counts))
print("Treino Balanceado")

df_train_ = pd.DataFrame.from_dict(hashtags_counts, orient='index')

plt.figure(figsize=(16, 10))
kde_kws={"color": "g", "alpha":0.3, "linewidth": 5, "shade":True }

sns.set()
ax = sns.distplot(df_train_, kde = False, vertical = True)
ax.grid()

```
